 ## Statistics Practical Implementation

### Measure Of Central Tendency

   1.Mean
   2.Median
   3.Mode
   

```python
ages = [23, 24, 32, 45, 12, 43, 67, 45, 32,56, 32,140]
```

```python
import numpy as np
print(np.mean(ages))
print(np.median(ages))
```

 ### We can find measure of central tendency using statistics

```python
import statistics
print(statistics.mode(ages))
print(statistics.mean(ages))
print(statistics.median(ages))
```

### Seaborn and boxplot

```python
import seaborn as sns
sns.boxplot(ages)
sns.displot(ages)
```

## 5 number summary

```python
import numpy as np
```

```python
q1 = np.percentile(ages,25)
q3 = np.percentile(ages,75)
```

```python
q1,q3
```

## Lower Fence and Upper Fence to check outliers


```python
iqr = q3-q1
lower_fence = q1-1.5*(iqr)
upper_fence = q3 +1.5*(iqr)
print(lower_fence, upper_fence)
```

## Measure OF Dispersion
1. Variance
2. Standard Deviation

```python
statistics.pvariance(ages)   # Gives Sample Variance
```

```python
np.var(ages)          # Gives Population Variance
```

## Check whether the distribution is Normal distributed 

```python
#### To check if the feature is gaussian or normal distributed 
## Q-Q plot
import matplotlib.pyplot as plt
import scipy.stats as stat
import pylab 
```

```python
def plot_data(sample):
    plt.figure(figsize=(10,6))
    plt.subplot(1,2,1)
    sns.histplot(sample)
    plt.subplot(1,2,2)
    stat.probplot(np.log(sample), dist='norm', plot=pylab)
    plt.show()
```

```python
## Create a normal distributed data
s= np.random.normal(0.5, 0.2, 1000)
plot_data(s)
```

```python
plot_data(df['sepal_width'])
```
